### Extracting Text of PDF Pages

```python
with filepdf.open('rb') as fp:
    pdf_reader = PyPDF2.PdfFileReader(fp)
    
    # printing first page contents
    pdf_page = pdf_reader.getPage(0)
    print(pdf_page.extractText())
    
    # reading all the pages content one by one
    for page_num in range(pdf_reader.numPages):
        pdf_page = pdf_reader.getPage(page_num)
        print(pdf_page.extractText())
```

 ### Rotate PDF File Pages

```python
filepdf2 = filepdf.with_name('Python_Tutorial_rotated.pdf')

with filepdf.open('rb') as fp:
    pdf_reader = PyPDF2.PdfFileReader(fp)
    pdf_writer = PyPDF2.PdfFileWriter()

    for page_num in range(pdf_reader.numPages):
        pdf_page = pdf_reader.getPage(page_num)
        pdf_page.rotateClockwise(90)  # rotateCounterClockwise()

        pdf_writer.addPage(pdf_page)

    with filepdf2.open('wb') as pdf_file_rotated:
        pdf_writer.write(pdf_file_rotated)
        print(pdf_file_rotated.name)
    
```

## Split PDF Files into Single Pages Files

```python
with filepdf.open('rb') as pdf_file:
    pdf_reader = PyPDF2.PdfFileReader(pdf_file)
    for i in range(pdf_reader.numPages):
        pdf_writer = PyPDF2.PdfFileWriter()
        pdf_writer.addPage(pdf_reader.getPage(i))
        output_file_name = Path(gettempdir()).joinpath('split', f'{filepdf.stem}_{i:0>3}.pdf')
        output_file_name.parent.mkdir(exist_ok=True, parents=True)
        with open(output_file_name, 'wb') as output_file:
            pdf_writer.write(output_file)
            print(output_file.name)
```
